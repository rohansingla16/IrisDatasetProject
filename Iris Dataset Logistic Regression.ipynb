```python
num_train_iris=65
data_train=iris[:num_train_iris]
data_test=iris[num_train_iris:]
```

```python
model=LogisticRegression()
```

```python
iris
```

```python
len(data_train)
```

```python
model.fit(data_train[['sepal_length','sepal_width']],data_train['category'])
```

```python
pred_train=model.predict(data_train[['sepal_length','sepal_width']])
```

```python
accuracy_train=(pred_train==data_train['category']).mean()
accuracy_train
```

```python
train=model.predict_proba(data_train[['sepal_length','sepal_width']])[:,1]
```

```python
plt.scatter(data_train['sepal_length'],train,marker='+',color='g')
plt.xlabel='sepal_length'
plt.ylabel='Prob'
print(plt.show())
```

```python
pred_test=model.predict(data_test[['sepal_length','sepal_width']])
```
